# Критерий знаков

```python
import pandas as pd
from statsmodels.stats.descriptivestats import sign_test
```

### Считываем данные

```python
data = pd.read_csv('data_big.csv')
data.head()
```

### Создаём нужный датафрейм

```python
df = pd.DataFrame()
df['anxiety.1'] = data['anxiety.1']
df['anxiety.2'] = data['anxiety.2']
df['result'] = data['anxiety.2'] - data['anxiety.1']
df
```

### Гипотезы

Нужно проверить значимость изменений в первые два дня ААС  
H_0: низкая значимость изменений в первые два дня ААС  
H_1: высокая значимость изменений в первые два дня ААС  
Пусть уровень значимости alpha = 0.005

```python
print("p-value =", sign_test(df['result'])[1])
```

p-value < alpha, тогда нулевая гипотеза отвергается  
В таком случае, мы отвергает тот факт, что в первые два дня была низкая значимость
