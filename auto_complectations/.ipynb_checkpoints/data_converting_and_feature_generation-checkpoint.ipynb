### Смотрим на уникальные значения

```python
def show_unique_data(data):
    for i in range(data.shape[1]):
        print('column name: ', data.columns[i]) 
        print('number of unique: ', data[data.columns[i]].unique().size) 
        print(data[data.columns[i]].unique())
        print()
```

```python
show_unique_data(data)
```

```python
# Вывод всех данных (если без break)
for i, row in data.iterrows():
    print(f"Index: {i}")
    print(f"{row}\n")
    break
```

### Смотрим, где есть пропуски в данных

```python
def no_info_in_data(data):
    for col_name, dt in data.items():
        for i in range(data.shape[0]):
            if dt[i] == 'нет данных':
                #print(data.iloc[i])
                print(i, col_name)
```

```python
no_info_in_data(data)
```

## Преобразование данных (работа с категориальными данными)

### Привод

```python
data['Привод'] = str('привод_') + data['Привод']
data = pd.concat([data, pd.get_dummies(data['Привод'])], axis=1).drop('Привод', axis=1)
data.head()
```

### Комплектация

Удаляем столбец, т.к. он скорее ухудшит работу алгоритма, чем улчучшит

```python
del data['complect']
```

### Модель

```python
data['model'] = str('model ') + data['model']
data = pd.concat([data, pd.get_dummies(data['model'])], axis=1).drop('model', axis=1)
data.head()
```

### Бренд

```python
data['brand'] = str('brand ') + data['brand']
data = pd.concat([data, pd.get_dummies(data['brand'])], axis=1).drop('brand', axis=1)
data.head()
```

### Тип двигателя

```python
data['тип_двигателя'] = str('тип_двигателя ') + data['тип_двигателя']
data = pd.concat([data, pd.get_dummies(data['тип_двигателя'])], axis=1).drop('тип_двигателя', axis=1)
data.head()
```

### Тип коробки передач

```python
data['тип_коробки_передач'] = str('тип_коробки_передач ') + data['тип_коробки_передач']
data = pd.concat([data, pd.get_dummies(data['тип_коробки_передач'])], axis=1).drop('тип_коробки_передач', axis=1)
data.head()
```
